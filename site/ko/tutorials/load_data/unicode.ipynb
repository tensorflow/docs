## ìœ ë‹ˆì½”ë“œ í‘œí˜„

í…ì„œí”Œë¡œì—ì„œ ìœ ë‹ˆì½”ë“œ ë¬¸ìì—´ì„ í‘œí˜„í•˜ê¸° ìœ„í•œ ë‘ ê°€ì§€ ë°©ë²•ì´ ìˆìŠµë‹ˆë‹¤:

* `string` ìŠ¤ì¹¼ë¼ â€”  ì½”ë“œ í¬ì¸íŠ¸ì˜ ì‹œí€€ìŠ¤ê°€ ì•Œë ¤ì§„ [ë¬¸ì ì¸ì½”ë”©](https://ko.wikipedia.org/wiki/%EB%AC%B8%EC%9E%90_%EC%9D%B8%EC%BD%94%EB%94%A9)ì„ ì‚¬ìš©í•´ ì¸ì½”ë”©ë©ë‹ˆë‹¤.
* `int32` ë²¡í„° â€” ìœ„ì¹˜ë§ˆë‹¤ ê°œë³„ ì½”ë“œ í¬ì¸íŠ¸ë¥¼ í¬í•¨í•©ë‹ˆë‹¤.

ì˜ˆë¥¼ ë“¤ì–´,  ì•„ë˜ì˜ ì„¸ ê°€ì§€ ê°’ì´ ëª¨ë‘ ìœ ë‹ˆì½”ë“œ ë¬¸ìì—´ `"è¯­è¨€å¤„ç†"`(ì¤‘êµ­ì–´ë¡œ "ì–¸ì–´ ì²˜ë¦¬"ë¥¼ ì˜ë¯¸í•¨)ë¥¼ í‘œí˜„í•©ë‹ˆë‹¤.

```python
# UTF-8ë¡œ ì¸ì½”ë”©ëœ string ìŠ¤ì¹¼ë¼ë¡œ í‘œí˜„í•œ ìœ ë‹ˆì½”ë“œ ë¬¸ìì—´ì…ë‹ˆë‹¤.
text_utf8 = tf.constant(u"è¯­è¨€å¤„ç†")
text_utf8
```

```python
# UTF-16-BEë¡œ ì¸ì½”ë”©ëœ string ìŠ¤ì¹¼ë¼ë¡œ í‘œí˜„í•œ ìœ ë‹ˆì½”ë“œ ë¬¸ìì—´ì…ë‹ˆë‹¤.
text_utf16be = tf.constant(u"è¯­è¨€å¤„ç†".encode("UTF-16-BE"))
text_utf16be
```

```python
# ìœ ë‹ˆì½”ë“œ ì½”ë“œ í¬ì¸íŠ¸ì˜ ë²¡í„°ë¡œ í‘œí˜„í•œ ìœ ë‹ˆì½”ë“œ ë¬¸ìì—´ì…ë‹ˆë‹¤.
text_chars = tf.constant([ord(char) for char in u"è¯­è¨€å¤„ç†"])
text_chars
```

### í‘œí˜„ ê°„ì˜ ë³€í™˜

í…ì„œí”Œë¡œëŠ” ë‹¤ë¥¸ í‘œí˜„ìœ¼ë¡œ ë³€í™˜í•˜ê¸° ìœ„í•œ ì—°ì‚°ì„ ì œê³µí•©ë‹ˆë‹¤.

* `tf.strings.unicode_decode`: ì¸ì½”ë”©ëœ `string` ìŠ¤ì¹¼ë¼ë¥¼ ì½”ë“œ í¬ì¸íŠ¸ì˜ ë²¡í„°ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
* `tf.strings.unicode_encode`: ì½”ë“œ í¬ì¸íŠ¸ì˜ ë²¡í„°ë¥¼ ì¸ì½”ë“œëœ `string` ìŠ¤ì¹¼ë¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
* `tf.strings.unicode_transcode`: ì¸ì½”ë“œëœ `string` ìŠ¤ì¹¼ë¼ë¥¼ ë‹¤ë¥¸ ì¸ì½”ë”©ìœ¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.

### ë°°ì¹˜(batch) ì°¨ì›

ì—¬ëŸ¬ ê°œì˜ ë¬¸ìì—´ì„ ë””ì½”ë”© í•  ë•Œ ë¬¸ìì—´ë§ˆë‹¤ í¬í•¨ëœ ë¬¸ìì˜ ê°œìˆ˜ëŠ” ë™ì¼í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤. ë°˜í™˜ë˜ëŠ” ê°’ì€ [`tf.RaggedTensor`](../../guide/ragged_tensor.ipynb)ë¡œ ê°€ì¥ ì•ˆìª½ ì°¨ì›ì˜ í¬ê¸°ê°€ ë¬¸ìì—´ì— í¬í•¨ëœ ë¬¸ìì˜ ê°œìˆ˜ì— ë”°ë¼ ê²°ì •ë©ë‹ˆë‹¤.

```python
# UTF-8 ì¸ì½”ë”©ëœ ë¬¸ìì—´ë¡œ í‘œí˜„í•œ ìœ ë‹ˆì½”ë“œ ë¬¸ìì—´ì˜ ë°°ì¹˜ì…ë‹ˆë‹¤. 
batch_utf8 = [s.encode('UTF-8') for s in
              [u'hÃƒllo',  u'What is the weather tomorrow',  u'GÃ¶Ã¶dnight', u'ğŸ˜Š']]
batch_chars_ragged = tf.strings.unicode_decode(batch_utf8,
                                               input_encoding='UTF-8')
for sentence_chars in batch_chars_ragged.to_list():
    print(sentence_chars)
```

`tf.RaggedTensor`ë¥¼ ë°”ë¡œ ì‚¬ìš©í•˜ê±°ë‚˜, íŒ¨ë”©(padding)ì„ ì‚¬ìš©í•´ `tf.Tensor`ë¡œ ë³€í™˜í•˜ê±°ë‚˜,  `tf.RaggedTensor.to_tensor` ì™€ `tf.RaggedTensor.to_sparse` ë©”ì„œë“œë¥¼ ì‚¬ìš©í•´ `tf.SparseTensor`ë¡œ ë³€í™˜í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
batch_chars_padded = batch_chars_ragged.to_tensor(default_value=-1)
print(batch_chars_padded.numpy())
```

```python
batch_chars_sparse = batch_chars_ragged.to_sparse()
```

ê¸¸ì´ê°€ ê°™ì€ ì—¬ëŸ¬ ë¬¸ìì—´ì„ ì¸ì½”ë”©í•  ë•ŒëŠ” `tf.Tensor`ë¥¼ ì…ë ¥ìœ¼ë¡œ ì‚¬ìš©í•©ë‹ˆë‹¤.

```python
tf.strings.unicode_encode([[99, 97, 116], [100, 111, 103], [ 99, 111, 119]],
                          output_encoding='UTF-8')
```

ê¸¸ì´ê°€ ë‹¤ë¥¸ ì—¬ëŸ¬ ë¬¸ìì—´ì„ ì¸ì½”ë”©í•  ë•ŒëŠ” `tf.RaggedTensor`ë¥¼ ì…ë ¥ìœ¼ë¡œ ì‚¬ìš©í•´ì•¼ í•©ë‹ˆë‹¤.

```python
tf.strings.unicode_encode(batch_chars_ragged, output_encoding='UTF-8')
```

íŒ¨ë”©ëœ í…ì„œë‚˜ í¬ì†Œ(sparse) í…ì„œëŠ” `unicode_encode`ë¥¼ í˜¸ì¶œí•˜ê¸° ì „ì— `tf.RaggedTensor`ë¡œ ë°”ê¿‰ë‹ˆë‹¤.

```python
tf.strings.unicode_encode(
    tf.RaggedTensor.from_sparse(batch_chars_sparse),
    output_encoding='UTF-8')
```

```python
tf.strings.unicode_encode(
    tf.RaggedTensor.from_tensor(batch_chars_padded, padding=-1),
    output_encoding='UTF-8')
```

### ê¸¸ì´

`tf.strings.length` ì—°ì‚°ì€ ê³„ì‚°í•´ì•¼ í•  ê¸¸ì´ë¥¼ ë‚˜íƒ€ë‚´ëŠ” `unit` ì¸ìë¥¼ ê°€ì§‘ë‹ˆë‹¤. `unit`ì˜ ê¸°ë³¸ ë‹¨ìœ„ëŠ” `"BYTE"`ì´ì§€ë§Œ ì¸ì½”ë”©ëœ `string`ì— í¬í•¨ëœ ìœ ë‹ˆì½”ë“œ ì½”ë“œ í¬ì¸íŠ¸ì˜ ìˆ˜ë¥¼ íŒŒì•…í•˜ê¸° ìœ„í•´ `"UTF8_CHAR"`ë‚˜ `"UTF16_CHAR"`ê°™ì´ ë‹¤ë¥¸ ê°’ì„ ì„¤ì •í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
# UTF8ì—ì„œ ë§ˆì§€ë§‰ ë¬¸ìëŠ” 4ë°”ì´íŠ¸ë¥¼ ì°¨ì§€í•©ë‹ˆë‹¤.
thanks = u'Thanks ğŸ˜Š'.encode('UTF-8')
num_bytes = tf.strings.length(thanks).numpy()
num_chars = tf.strings.length(thanks, unit='UTF8_CHAR').numpy()
print('{} ë°”ì´íŠ¸; {}ê°œì˜ UTF-8 ë¬¸ì'.format(num_bytes, num_chars))
```

### ë¶€ë¶„ ë¬¸ìì—´

ì´ì™€ ìœ ì‚¬í•˜ê²Œ `tf.strings.substr` ì—°ì‚°ì€  "`unit`" ë§¤ê°œë³€ìˆ˜ ê°’ì„ ì‚¬ìš©í•´ "`pos`"ì™€ "`len`" ë§¤ê°œë³€ìˆ˜ë¡œ ì§€ì •ëœ ë¬¸ìì—´ì˜ ì¢…ë¥˜ë¥¼ ê²°ì •í•©ë‹ˆë‹¤.

### ë¬¸ì ë°”ì´íŠ¸ ì˜¤í”„ì…‹

`tf.strings.unicode_decode`ë¡œ ë§Œë“  ë¬¸ì í…ì„œë¥¼ ì›ë³¸ ë¬¸ìì—´ê³¼ ìœ„ì¹˜ë¥¼ ë§ì¶”ë ¤ë©´ ê° ë¬¸ìì˜ ì‹œì‘ ìœ„ì¹˜ì˜ ì˜¤í”„ì…‹(offset)ì„ ì•Œì•„ì•¼ í•©ë‹ˆë‹¤. `tf.strings.unicode_decode_with_offsets`ì€ `unicode_decode`ì™€ ë¹„ìŠ·í•˜ì§€ë§Œ ê° ë¬¸ìì˜ ì‹œì‘ ì˜¤í”„ì…‹ì„ í¬í•¨í•œ ë‘ ë²ˆì§¸ í…ì„œë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.

## ìœ ë‹ˆì½”ë“œ ìŠ¤í¬ë¦½íŠ¸

ê° ìœ ë‹ˆì½”ë“œ ì½”ë“œ í¬ì¸íŠ¸ëŠ” [ìŠ¤í¬ë¦½íŠ¸](https://en.wikipedia.org/wiki/Script_%28Unicode%29)(script)ë¼ ë¶€ë¥´ëŠ” í•˜ë‚˜ì˜ ì½”ë“œ í¬ì¸íŠ¸ì˜ ì§‘í•©(collection)ì— ì†í•©ë‹ˆë‹¤. ë¬¸ìì˜ ìŠ¤í¬ë¦½íŠ¸ëŠ” ë¬¸ìê°€ ì–´ë–¤ ì–¸ì–´ì¸ì§€ ê²°ì •í•˜ëŠ” ë° ë„ì›€ì´ ë©ë‹ˆë‹¤. ì˜ˆë¥¼ ë“¤ì–´, 'Ğ‘'ê°€ í‚¤ë¦´(Cyrillic) ìŠ¤í¬ë¦½íŠ¸ë¼ëŠ” ê²ƒì„ ì•Œê³  ìˆìœ¼ë©´ ì´ ë¬¸ìê°€ í¬í•¨ëœ í…ìŠ¤íŠ¸ëŠ” ì•„ë§ˆë„ (ëŸ¬ì‹œì•„ì–´ë‚˜ ìš°í¬ë¼ì´ë‚˜ì–´ ê°™ì€) ìŠ¬ë¼ë¸Œ ì–¸ì–´ë¼ëŠ” ê²ƒì„ ì•Œ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ìŠ¤í¬ë¦½íŠ¸ì˜ ë³€í™”ë¥¼ ë‹¨ì–´ ê²½ê³„ë¡œ ê·¼ì‚¬í•˜ì—¬ (ML ëª¨ë¸ ì‚¬ìš© ì—†ì´) ëŒ€ëµì ì¸ ë¶„í• ì„ ìˆ˜í–‰í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ìœ„ì—ì„œ ì–¸ê¸‰ëœ "NYæ ªä¾¡"ì˜ ì˜ˆì™€ ê°™ì€ ë¬¸ìì—´ì— ì ìš©ë©ë‹ˆë‹¤. ë‹¤ì–‘í•œ ìŠ¤í¬ë¦½íŠ¸ì˜ ê³µë°± ë¬¸ìë¥¼ ëª¨ë‘ USCRIPT_COMMON(ì‹¤ì œ í…ìŠ¤íŠ¸ì˜ ìŠ¤í¬ë¦½íŠ¸ ì½”ë“œì™€ ë‹¤ë¥¸ íŠ¹ë³„í•œ ìŠ¤í¬ë¦½íŠ¸ ì½”ë“œ)ìœ¼ë¡œ ë¶„ë¥˜í•˜ê¸° ë•Œë¬¸ì— ê³µë°±ì„ ì‚¬ìš©í•˜ëŠ” ëŒ€ë¶€ë¶„ì˜ ì–¸ì–´ë“¤ì—ì„œë„ ì—­ì‹œ ì ìš©ë©ë‹ˆë‹¤. 

```python
# dtype: string; shape: [num_sentences]
#
# ì²˜ë¦¬í•  ë¬¸ì¥ë“¤ ì…ë‹ˆë‹¤. ì´ ë¼ì¸ì„ ìˆ˜ì •í•´ì„œ ë‹¤ë¥¸ ì…ë ¥ê°’ì„ ì‹œë„í•´ ë³´ì„¸ìš”!
sentence_texts = [u'Hello, world.', u'ä¸–ç•Œã“ã‚“ã«ã¡ã¯']
```

ë¨¼ì € ë¬¸ì¥ì„ ë¬¸ì ì½”ë“œ í¬ì¸íŠ¸ë¡œ ë””ì½”ë”©í•˜ê³  ê° ë¬¸ìì— ëŒ€í•œ ìŠ¤í¬ë¦½íŠ¸ ì‹ë³„ìë¥¼ ì°¾ìŠµë‹ˆë‹¤.

```python
# dtype: int32; shape: [num_sentences, (num_chars_per_sentence)]
#
# sentence_char_codepoint[i, j]ëŠ”
# ië²ˆì§¸ ë¬¸ì¥ ì•ˆì— ìˆëŠ” jë²ˆì§¸ ë¬¸ìì— ëŒ€í•œ ì½”ë“œ í¬ì¸íŠ¸ ì…ë‹ˆë‹¤.
sentence_char_codepoint = tf.strings.unicode_decode(sentence_texts, 'UTF-8')
print(sentence_char_codepoint)

# dtype: int32; shape: [num_sentences, (num_chars_per_sentence)]
#
# sentence_char_codepoint[i, j]ëŠ” 
# ië²ˆì§¸ ë¬¸ì¥ ì•ˆì— ìˆëŠ” jë²ˆì§¸ ë¬¸ìì˜ ìœ ë‹ˆì½”ë“œ ìŠ¤í¬ë¦½íŠ¸ ì…ë‹ˆë‹¤.
sentence_char_script = tf.strings.unicode_script(sentence_char_codepoint)
print(sentence_char_script)
```

ê·¸ë‹¤ìŒ ìŠ¤í¬ë¦½íŠ¸ ì‹ë³„ìë¥¼ ì‚¬ìš©í•˜ì—¬ ë‹¨ì–´ ê²½ê³„ê°€ ì¶”ê°€ë  ìœ„ì¹˜ë¥¼ ê²°ì •í•©ë‹ˆë‹¤. ê° ë¬¸ì¥ì˜ ì‹œì‘ê³¼ ì´ì „ ë¬¸ìì™€ ìŠ¤í¬ë¦½íŠ¸ê°€ ë‹¤ë¥¸ ë¬¸ìì— ë‹¨ì–´ ê²½ê³„ë¥¼ ì¶”ê°€í•©ë‹ˆë‹¤.

```python
# dtype: bool; shape: [num_sentences, (num_chars_per_sentence)]
#
# sentence_char_starts_word[i, j]ëŠ” 
# ië²ˆì§¸ ë¬¸ì¥ ì•ˆì— ìˆëŠ” jë²ˆì§¸ ë¬¸ìê°€ ë‹¨ì–´ì˜ ì‹œì‘ì´ë©´ True ì…ë‹ˆë‹¤.
sentence_char_starts_word = tf.concat(
    [tf.fill([sentence_char_script.nrows(), 1], True),
     tf.not_equal(sentence_char_script[:, 1:], sentence_char_script[:, :-1])],
    axis=1)

# dtype: int64; shape: [num_words]
#
# word_starts[i]ì€ (ëª¨ë“  ë¬¸ì¥ì˜ ë¬¸ìë¥¼ ì¼ë ¬ë¡œ í¼ì¹œ ë¦¬ìŠ¤íŠ¸ì—ì„œ)
# ië²ˆì§¸ ë‹¨ì–´ê°€ ì‹œì‘ë˜ëŠ” ë¬¸ìì˜ ì¸ë±ìŠ¤ ì…ë‹ˆë‹¤.
word_starts = tf.squeeze(tf.where(sentence_char_starts_word.values), axis=1)
print(word_starts)
```

ì´ ì‹œì‘ ì˜¤í”„ì…‹ì„ ì‚¬ìš©í•˜ì—¬ ì „ì²´ ë°°ì¹˜ì— ìˆëŠ” ë‹¨ì–´ ë¦¬ìŠ¤íŠ¸ë¥¼ ë‹´ì€ `RaggedTensor`ë¥¼ ë§Œë“­ë‹ˆë‹¤.

```python
# dtype: int32; shape: [num_words, (num_chars_per_word)]
#
# word_char_codepoint[i, j]ì€ 
# ië²ˆì§¸ ë‹¨ì–´ ì•ˆì— ìˆëŠ” jë²ˆì§¸ ë¬¸ìì— ëŒ€í•œ ì½”ë“œ í¬ì¸íŠ¸ ì…ë‹ˆë‹¤.
word_char_codepoint = tf.RaggedTensor.from_row_starts(
    values=sentence_char_codepoint.values,
    row_starts=word_starts)
print(word_char_codepoint)
```

ë§ˆì§€ë§‰ìœ¼ë¡œ ë‹¨ì–´ ì½”ë“œ í¬ì¸íŠ¸ `RaggedTensor`ë¥¼ ë¬¸ì¥ìœ¼ë¡œ ë‹¤ì‹œ ë‚˜ëˆ•ë‹ˆë‹¤.

```python
# dtype: int64; shape: [num_sentences]
#
# sentence_num_words[i]ëŠ” ië²ˆì§¸ ë¬¸ì¥ ì•ˆì— ìˆëŠ” ë‹¨ì–´ì˜ ìˆ˜ì…ë‹ˆë‹¤.
sentence_num_words = tf.reduce_sum(
    tf.cast(sentence_char_starts_word, tf.int64),
    axis=1)

# dtype: int32; shape: [num_sentences, (num_words_per_sentence), (num_chars_per_word)]
#
# sentence_word_char_codepoint[i, j, k]ëŠ” ië²ˆì§¸ ë¬¸ì¥ ì•ˆì— ìˆëŠ”
# jë²ˆì§¸ ë‹¨ì–´ ì•ˆì˜ kë²ˆì§¸ ë¬¸ìì— ëŒ€í•œ ì½”ë“œ í¬ì¸íŠ¸ì…ë‹ˆë‹¤.
sentence_word_char_codepoint = tf.RaggedTensor.from_row_lengths(
    values=word_char_codepoint,
    row_lengths=sentence_num_words)
print(sentence_word_char_codepoint)
```
